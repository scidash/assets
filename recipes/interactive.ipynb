```python
%matplotlib inline
import ipywidgets as iw
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(z, c):
    x = np.linspace(0,10,1000)
    y = x**z
    maxx = y.max()
    y /= maxx
    color = 'gray'
    plt.plot(x,y**2,c=(c if c=='red' else color))
    plt.plot(x,y,c=(c if c=='green' else color))
    plt.plot(x,y**0.5,c=(c if c=='blue' else color))
```

```python
iw.interact(f,
         z=iw.widgets.IntSlider(min=1,max=30,step=1,value=10),
         c=iw.RadioButtons(options=['blue', 'green', 'red'],disabled=False,description='Color'));
```

```python
def g(z,):
    fig, ax1 = plt.subplots()

    # These are in unitless percentages of the figure size. (0,0 is bottom left)
    left, bottom, width, height = [0.25, 0.6, 0.2, 0.2]
    ax2 = fig.add_axes([left, bottom, width, height])

    ax1.plot(range(10), color='red')
    ax1.scatter([z],[z],s=20)
    for i in range(0,10):
        x = np.arange(6)
        ax2.plot(x,x*i,color='gray')
    ax2.plot(x,x*z,color='red');
    
iw.interact(g,
         z=iw.widgets.IntSlider(min=0,max=9,step=1,value=3));
```
